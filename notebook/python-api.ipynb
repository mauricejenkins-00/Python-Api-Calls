```python
# Importing Dependencies 

import pandas as pd
import requests
import time
import pprint
import json
from pandas import json_normalize
```

```python
# Declaring Variables 

key = "AIzaSyCDjfrvTicv7OQopYJWL4o7GiwmXQdbKHM"
places_url = f"https://maps.googleapis.com/maps/api/place/nearbysearch/json?&key={key}"

data = ["","",""]

places = ["41.611, -87.6348", "41.6156, -87.5295", "41.5039, -87.7131"]
```

```python
# Placing a request to the Google Places Api and storing the returned JSON data in our empty data list

for i in places: 
    params = {
        "location": i,
        "radius": 50000,
        "keyword": "restaurants"
    }

response = requests.get(places_url, params=params).json()
data.insert(i, response)
```

```python
# Pretty Printing a json response 

pprint.pprint(data)
```

```python
# Convert this json response into a dataframe 

phoenix_df = pd.json_normalize(response, record_path="results")
phoenix_df = phoenix_df[["name", "business_status", "vicinity", "user_ratings_total", "types"]]
phoenix_df.head()
```
